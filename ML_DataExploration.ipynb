```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv("LinearPlot.csv")

X = df[['input_column']]   # independent variable
y = df['target_column']    # dependent variable
```

```python
plt.scatter(X, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter Plot")
plt.show()
```

```python
corr = np.corrcoef(df['input_column'], df['target_column'])[0,1]
print("Correlation:", corr)
```

```python
lin_model = LinearRegression()
lin_model.fit(X, y)

y_pred_lin = lin_model.predict(X)

print("Linear R2:", r2_score(y, y_pred_lin))
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred_lin, color='red')
plt.title("Linear Fit")
plt.show()
```

```python
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

poly_model = LinearRegression()
poly_model.fit(X_poly, y)

y_pred_poly = poly_model.predict(X_poly)

print("Polynomial R2:", r2_score(y, y_pred_poly))
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred_poly, color='green')
plt.title("Polynomial Fit")
plt.show()
```

```python
if abs(corr) > 0.7:
    print("Strong Linear Relationship")
elif r2_score(y, y_pred_poly) > r2_score(y, y_pred_lin):
    print("Non-linear (Polynomial fits better)")
else:
    print("Weak or No clear relationship")
```
